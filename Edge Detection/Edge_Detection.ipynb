```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
import math
```

```python
PATH = '/content/drive/MyDrive/Study Documents/CS231 - CV/CS231.L21.KHTN/EdgeDetection/peppers.jpg'
```

```python
def edgeDectect(path, num):
  img = cv2.imread(path, 0)
  kernelsX = np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
  kernelsY = np.array([[-1,-2,-1],[0,0,0],[1,2,1]])
  kernelsD1 = np.array([[-2,-1,0],[-1,0,1],[0,1,2]])
  kernelsD2 = np.array([[0,1,2],[-1,0,1],[-2,-1,0]])
  if num == 1:
    return cv2.filter2D(img, -1, kernelsX)
  elif num == 2:
    return cv2.filter2D(img, -1, kernelsY)
  elif num == 3:
    return cv2.filter2D(img, -1, kernelsD1)
  else:
    return cv2.filter2D(img, -1, kernelsD2)
```

```python
cv2_imshow(img)
```

```python
cv2_imshow(edgeDectect(PATH,1))
```

```python
def sumConv(mat1, mat2):
  matSum = mat1
  for i in range(mat1.shape[0]):
    for j in range(mat1.shape[1]):
      matSum[i][j] = math.sqrt(mat1[i][j]**2 + mat2[i][j]**2)
  return matSum
```

```python
imgRst1 = edgeDectect(PATH, 1)
imgRst2 = edgeDectect(PATH, 2)
imgSum = sumConv(imgRst1, imgRst2)
cv2_imshow(imgSum)
```
